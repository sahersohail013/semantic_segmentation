**SPLITTING THE DATA INTO TRAIN AND VALIDATION**

```python
import os
import numpy as np
import shutil

# # Creating Train / Val / Test folders (One time use)
root_dir = '/content/A3_Dataset'

os.makedirs(root_dir +'/images_prepped_train' )
os.makedirs(root_dir +'/images_prepped_val' )

# Creating partitions of the data after shuffeling

src = "/content/A3_Dataset/dataset1/images_prepped_train" # Folder to copy images from

allFileNames = os.listdir(src)
np.random.shuffle(allFileNames)

val_ratio=0.20

train_FileNames, val_FileNames = np.split(np.array(allFileNames),[int(len(allFileNames)* (1 - val_ratio))])



train_FileNames = [src+'/'+ name for name in train_FileNames.tolist()]
val_FileNames = [src+'/' + name for name in val_FileNames.tolist()]


print('Total images: ', len(allFileNames))
print('Training: ', len(train_FileNames))
print('Validation: ', len(val_FileNames))


# Copy-pasting images
for name in train_FileNames:
    shutil.copy(name, "/content/dataset1/images_prepped_train")

for name in val_FileNames:
    shutil.copy(name, "/content/dataset1/images_prepped_train")
```

```python
import os
import numpy as np
import shutil

# # Creating Train / Val / Test folders (One time use)
root_dir = '/content/A3_Dataset'

os.makedirs(root_dir +'/annotated_prepped_train' )
os.makedirs(root_dir +'/annotated_prepped_val' )

# Creating partitions of the data after shuffeling

src = "/content/A3_Dataset/dataset1/annotations_prepped_train" # Folder to copy images from

allFileNames = os.listdir(src)
np.random.shuffle(allFileNames)

val_ratio=0.20

train_FileNames, val_FileNames = np.split(np.array(allFileNames),[int(len(allFileNames)* (1 - val_ratio))])



train_FileNames = [src+'/'+ name for name in train_FileNames.tolist()]
val_FileNames = [src+'/' + name for name in val_FileNames.tolist()]


print('Total images: ', len(allFileNames))
print('Training: ', len(train_FileNames))
print('Validation: ', len(val_FileNames))


# Copy-pasting images
for name in train_FileNames:
    shutil.copy(name, "/content/A3_Dataset/annotated_prepped_train")

for name in val_FileNames:
    shutil.copy(name, "/content/A3_Dataset/annotated_prepped_val")
```

**MODEL FITTING**

RESNET50_UNET

```python
from tensorflow import keras

from keras_segmentation.models.unet import resnet50_unet

model = resnet50_unet(n_classes=12 ,  input_height=224, input_width=224  )
```

```python
model.train(
    train_images =  "/content/dataset1/images_prepped_train",
    train_annotations = "/content/dataset1/annotations_prepped_train",
    input_height=224,
    input_width=224,
    n_classes=12,
    do_augment=True,
    val_images="/content/dataset1/images_prepped_val",
    val_annotations="/content/dataset1/annotations_prepped_val",
    augmentation_name="aug_all",
    epochs=100,
    validate=True,
    batch_size=2,
    val_batch_size=2,
    steps_per_epoch=50,
    val_steps_per_epoch=35,
)
```

MODEL EVALUATION

```python
out = model.predict_segmentation(
    inp="/content/dataset1/images_prepped_test/0016E5_08155.png",
    out_fname="/tmp/out.png"
)

import matplotlib.pyplot as plt
plt.imshow(out)

# evaluating the model 
print(model.evaluate_segmentation( inp_images_dir="/content/dataset1/images_prepped_test"  , annotations_dir="/content/dataset1/annotations_prepped_test" ) )
```
